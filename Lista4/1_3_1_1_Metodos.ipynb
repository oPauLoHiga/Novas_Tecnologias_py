Já vimos alguns exemplos de métodos ao aprender sobre tipos de estrutura de objetos e dados em Python. Métodos são essencialmente funções construídas em objetos. Mais tarde, você verá que podemos pensar em métodos como tendo um argumento 'self' referente ao próprio objeto.

**Métodos (methods)**
Um método é, essencialmente, uma função que pertence a um objeto ou classe. Ele é definido dentro do corpo de uma classe e descreve um "comportamento" que aquele objeto pode ter.

**Dependência**: Você só consegue chamar um método através de uma instância (objeto).

**O parâmetro self**: Quase todo método recebe o self como primeiro argumento, que representa a própria instância do objeto que o está chamando.

**Exemplo clássico**: .append() (da lista), .upper() (da string), .readlines() (do arquivo).

```python
class Robo:
    def ligar(self): # Isso é um método
        print("Bip Bop! Robô ligado.")

meu_robo = Robo()
meu_robo.ligar() # Chamada via objeto
```

# **Característica	Função vs Método**

**Onde está?**

**Função**: Solta no script (Global)	          

**Método**: Dentro de uma classe

**Como chamar?**

**Função**: nome_funcao(dados)

**Método**: objeto.nome_metodo()

**Relação**

**Função**: Opera sobre os dados passados

**Método**: Opera sobre os dados do objeto (self)

**Exemplo**

**Função**: sum([1, 2])

**Método**: "texto".upper()
